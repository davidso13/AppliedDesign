```python
%matplotlib inline
from matplotlib import pyplot as plt # matplotlib library import

X_axis = ['8000', '12000', '16000', '24000']
Y_axis = [62828.25402, 38031.00168, 35901.85944, 26926.39458]

xs = [i for i, _ in enumerate(Y_axis)]
plt.xticks([i for i, _ in enumerate(X_axis)], X_axis)
plt.bar(xs, Y_axis, align = 'center', alpha = 0.5)

plt.title("PC Cost")
plt.xlabel('Length')
plt.ylabel('Cost')

plt.show()
```
