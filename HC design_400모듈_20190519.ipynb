```python
import math

hollow_type = int(input("Hollow type : 1.원형 2.직사각형 3.타원 4.Solid "))
h_b = int(input("Input the width(mm) : "))
h_h = int(input("Input the height(mm) : "))

if (hollow_type == 1) :
    if(h_b == h_h):
        Hollow_A = h_b/2 * h_h/2 * math.pi

elif (hollow_type == 2):
    Hollow_A = h_b * h_h
    
elif (hollow_type == 3):
    Hollow_A = h_b/2 * h_h/2 * math.pi
    
elif (hollow_type == 4):
    Hollow_A = 0
    
Amount_steel = 287 * 7  # 287 * 5, 287 * 7
Total_A = [400*300, 400*350, 400*400, 400*450, 400*500]
print("Total_A", Total_A)


ratio_of_hollow = [] # hollow ratio 계산
for i in Total_A:
    ratio = Hollow_A / i * 100
    ratio_of_hollow.append(ratio)
    
Ac = []
for i in Total_A:
    Ac_elem = i - Hollow_A
    Ac.append(Ac_elem)
    
ratio_of_rebar = []
for i in Ac:
    rate = Amount_steel / i
    ratio_of_rebar.append(rate)
    
print()
print("Amount steel : ", Amount_steel)
print("Hollow area : ", Hollow_A)
print("Hollow ratio : ", ratio_of_hollow, "(%)")
print("Area of con : ", Ac)
print("Ratio of rebar : ", ratio_of_rebar)
```

```python
#import math
print("Hollow Area : ", Hollow_A)
hollow_type = int(input("Hollow type : 1.원형 2.직사각형 3.타원 4. Solid"))

if (hollow_type == 1): # 원형인 경우
    rec_w = math.sqrt(Hollow_A)
    rec_h = math.sqrt(Hollow_A)
elif (hollow_type == 2): # 직사각형인 경우
    rec_w = h_b
    rec_h = h_h
elif(hollow_type == 3): # 타원인 경우
    rec_w = h_b/2 * math.sqrt(math.pi)
    rec_h = h_h/2 * math.sqrt(math.pi)
elif(hollow_type == 4): # 도넛인 경우
    rec_w = 0
    rec_h = 0

d = [300, 350, 400, 450, 500]
dt = [260, 310, 360, 410, 460]

tf = [] # flange의 두께를 저장
for i in d:
    tf_elem = (i - rec_h) / 2
    tf.append(tf_elem)
    
bw = 400 - rec_w

# C, M 계산
fck = 24.0
fy = 400.0
As = 287 * 7 # 5-D19

# 휨철근, 전단철근 계산
Asf = []
Asw = []
for i in tf:
    Asf_elem = 0.85 * fck * i *(400 - bw) / fy
    Asw_elem = As - Asf_elem
    Asf.append(Asf_elem)
    Asw.append(Asw_elem)

# 중립축 계산
aw = []
for i in Asw:
    aw_elem = i * fy / 0.85 / fck / bw
    aw.append(aw_elem)

# C!, C2 계산
C1 = []
for i in tf:
    C1_elem = 0.85 * fck * i *(400 - bw)
    C1.append(C1_elem)
C2 = []
for j in aw:
    C2_elem = 0.85 * fck * j * bw
    C2.append(C2_elem)


Mn1 = []
Mn2 = []
Mn = []
k = 0
while k < 5:
    Mn1_elem = Asf[k] * fy * (dt[k] - tf[k]/2)
    Mn2_elem = Asw[k] * fy * (dt[k] - aw[k]/2)
    Mn_elem = Mn1_elem + Mn2_elem
    Mn1.append(Mn1_elem)
    Mn2.append(Mn2_elem)
    Mn.append(Mn_elem)
    k += 1

reduction_factor = 0.85

piMn = []
for i in Mn:
    piMn_elem = i * reduction_factor
    piMn.append(piMn_elem)

print("Rec_width : ", rec_w, " mm")
print("Rec_height : ", rec_h, " mm")

print("tf : ", tf , " mm")
print("bw : ", bw, " mm")
print("Asf : ", Asf, "mm**2")
print("Asw : ", Asw, "mm**2")
print("aw : ", aw)
print()
print()
print("C1 : ", C1)
print("C2 : ", C2)
#print("C : ", C)
print("Mn1 : ", Mn1, "(N mm)")
#print("Mn1 : ", Mn1 / 10**6, "(kN m)")
print("Mn2 : ", Mn2, "(N mm)")
#print("Mn2 : ", Mn2 / 10**6, "(kN m)")
print("Mn : ", Mn, "(N mm)")
#print("Mn : ", Mn / 10**6, "(kN m)")
print("piMn : ", piMn, "(N mm)")
```

## 자중 계산

```python
import math

length = int(input("Enter the length of slab : "))
height = int(input("Enter the height of slab : "))
row = length / 400
col = 6000 / 400
hollow_num = row * col

h_type = int(input("Hollow type : 1.원형 2.직사각형 3.타원 4.Solid "))
Area = length * 6000
Volumn = Area * height

if h_type == 1 :
    V_h = 4/3 * math.pi * 70 * 70 * 70 * hollow_num

elif h_type == 2:
    V_h = 270*270*140 * hollow_num
    
elif h_type == 3:
    V_h = 4/3 * (270 / 2) * (270/2) * 140 /2 * math.pi * hollow_num

elif h_type == 4 :
    V_h = 0

V = Volumn - V_h # 중공체 부피 제외한 면적
ratio = V_h/Volumn
Total_w = V * 2400 / 1000000000 # mm3으로 바꿔줌
Self_w = Total_w / Area * 100

print(hollow_num)
print("Area = " , Area , " mm**2")
print("Volumn = ", Volumn , " mm**3")
print("V = ", V , " mm**3" )
print("Vh = ", V_h, " mm**3")
print("hollow ratio = ", ratio * 100)
print("Total weight = ", Total_w, " kg")
print("self_w = ", Self_w, " kg/cm**2")
```
